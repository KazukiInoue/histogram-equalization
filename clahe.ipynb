```python
%matplotlib inline
import numpy as np
import cv2

import matplotlib.pyplot as plt
```

```python
def cv2_to_pltshow(in_img):
    blue = in_img[:, :, 0]
    green = in_img[:, :, 1]
    red = in_img[:, :, 2]
    
    out_img = np.zeros_like(in_img)
    out_img[:, :, 0] = red
    out_img[:, :, 1] = green
    out_img[:, :, 2] = blue
    out_img = out_img * 255
    out_img = out_img.astype(np.uint8)
    
    return out_img
```

```python
# 
image = cv2.imread("./input_images/Gakki01.jpg")
image = image / 255.0

plt.imshow(cv2_to_pltshow(image))
plt.axis('off')
plt.show()
print(image)
```

```python
# 
noise = np.random.normal(size=image.shape, scale=0.2)
noisy = np.clip(image + noise, 0.0, 1.0)
plt.imshow(noisy)
plt.axis('off')
plt.show()
```
