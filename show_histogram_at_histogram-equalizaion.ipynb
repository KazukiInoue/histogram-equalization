```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def cv2gray_to_pltshow(gray_img):
    out_img = np.zeros((gray_img.shape[0], gray_img.shape[1], 3))
    out_img = out_img.astype(np.uint8)
    out_img[:, :, 0] = np.copy(gray_img)
    out_img[:, :, 1] = np.copy(gray_img)
    out_img[:, :, 2] = np.copy(gray_img)
    
    return out_img
```

```python
src = cv2.imread("piobasan.jpg", 0)
equ = cv2.equalizeHist(src)

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
ax0.imshow(cv2gray_to_pltshow(src))
ax0.set_title('src image')
ax0.axis('off')

ax1 = fig.add_subplot(1,2,2)
ax1.imshow(cv2gray_to_pltshow(equ))
ax1.set_title('after histogram equalization')
ax1.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax0 = fig.add_subplot(1,2,1)
ax0.hist(src.flatten(),256,[0,256], color = 'k')
ax0.set_xlim([0,256])
ax0.set_title('src image')

ax1 = fig.add_subplot(1,2,2)
ax1.hist(equ.flatten(),256,[0,256], color = 'r')
ax1.set_xlim([0,256])
ax1.set_title('after histogram equalization')
plt.show()
```
